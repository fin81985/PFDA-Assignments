```python
# Assignment 3 domains
# Author: Finian Doonan


import pandas as pd
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"



df = pd.read_csv(url)
print (df.head)
```

```python
plt.figure(figsize=(8, 8))

# compute top email domains from the dataframe
top_domains = df['Email'].str.split('@').str[-1].value_counts().head(10)

# select as many colors as needed from a colormap
colors = plt.cm.tab20.colors[:len(top_domains)]

plt.pie(
    top_domains,
    labels=top_domains.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)

plt.title('Pie Chart of top Email Domains')
plt.axis('equal') 
plt.show()
```
